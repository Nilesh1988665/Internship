```python
happy_df = df.rename({'Country':'country_name', 'Region':'Region_name', 'Happiness Rank':'Happiness_Rank', 'Happiness Score':'Happiness_Score', 'Economy (GDP per Capita)':'Economy_GDP', 'Health (Life Expectancy)':'Life_Expectancy'}, axis =1)
```

```python
happy_df.head()
```

```python
plt.rcParams['figure.figsize'] = (15, 7)
plt.title('plot Between Hapiness Score and GDP')
sns.scatterplot(x = happy_df.Happiness_Score, y = happy_df.Economy_GDP)

```

```python
gdp_region = happy_df.groupby('Region_name')['Economy_GDP'].sum()

```

```python
gdp_region
```

```python
gdp_region.plot.pie(autopct = '%1.1f%%')
plt.title('GDP per Region')
plt.ylable('')
```

```python
total_country = happy_df.groupby('Region_name')[['country_name']].count()
total_country
```

```python
cor = happy_df.corr(method = "pearson")
f, ax = plt.subplots(figsize = (10, 5))
sns.heatmap(cor, mask = np.zeros_like(cor, dtype=np.bool),cmap="Blues", square=True, ax=ax)

```

```python
top_10 = happy_df.head(10)
bottom_10 = happy_df.tail(10)
```

```python
fig, axes= plt.subplots(1,2, figsize= (16, 6))
plt.tight_layout(pad= 2)
xlabels= top_10.country_name
axes[0].set_title('top 10 happiest countries Life_Expectancy')
axes[0].set_xticklabels(xlabels, rotation=45, ha='right')
sns.barplot(x= top_10.country_name, y= top_10.Life_Expectancy, ax= axes[0])
axes[0].set_xlables('country_name')
axes[0].set_ylables('Life_Expectancy')


xlables= bottom_10.country_name
axes[0].set_title('bottom 10 happiest countries Life Expectancy')
axes[0].set_xticklabls(xlabels, rotation=45, ha='right')
sns.barplot(x= bottom_10.country_name, y= top_10.Life_Expectancy, ax= axes[0])
axes[0].set_xlables('country_name')
axes[0].set_ylables('Life_Expectancy')
```
